```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('results.csv')
df['Throughput'] = df['Bank region size (B)'] / df['Time (s)']

sns.lineplot(
    data=df,
    x='Bank region size (B)',
    y='Time (s)'
)
plt.xscale('log')

df.groupby('Bank region size (B)').agg({'Time (s)': 'mean'})['Time (s)']
```
